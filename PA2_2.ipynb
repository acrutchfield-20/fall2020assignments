```python
#I had to create a copy of the original data file to delete all data but US data
#the original file is too large to perform time series on
us_data = pd.read_csv('input/fossil-fuel-co2-emissions-by-nation_filteredcopy.csv')
us_data

```

```python
#time series for US data
colors = ['red', 'blue', 'green','orange', 'purple', 'black','yellow', 'pink']
circle_size = 2
circle_opacity = 0.5
regression_opacity = 1
regression_method = "poly" 
width = 550
height = 220

#Field ID 3
base3 = alt.Chart(us_data).mark_circle(opacity=circle_opacity, color=colors[0], size=circle_size).encode(
    alt.X('Year:Q'),
    alt.Y('Total:Q')
).properties(width=width, height=height)
line3 = base3.mark_line(color=colors[0], opacity=1,size=1).encode(x='Year:Q',y='Total:Q')

#Field ID 4
base4 = alt.Chart(us_data).mark_circle(opacity=circle_opacity, color=colors[1], size=circle_size).encode(
    alt.X('Year:Q'),
   alt.Y('Solid Fuel:Q')
).properties(width=width, height=height)
line4 = base4.mark_line(color=colors[1], opacity=1,size=1).encode(x='Year:Q',y='Solid Fuel:Q')

#Field ID 5
base5 = alt.Chart(us_data).mark_circle(opacity=circle_opacity, color=colors[2], size=circle_size).encode(
    alt.X('Year:Q'),
    alt.Y('Liquid Fuel:Q')
).properties(width=width, height=height)
line5 = base5.mark_line(color=colors[2], opacity=1,size=1).encode(x='Year:Q',y='Liquid Fuel:Q')

#Field ID 6
base6 = alt.Chart(us_data).mark_circle(opacity=circle_opacity, color=colors[3], size=circle_size).encode(
    alt.X('Year:Q'),
    alt.Y('Gas Fuel:Q')
).properties(width=width, height=height)
line6 = base6.mark_line(color=colors[3], opacity=1,size=1).encode(x='Year:Q',y='Gas Fuel:Q')

#Field ID 7
base7 = alt.Chart(us_data).mark_circle(opacity=circle_opacity, color=colors[4], size=circle_size).encode(
    alt.X('Year:Q'),
    alt.Y('Cement:Q')
).properties(width=width, height=height)
line7 = base7.mark_line(color=colors[4], opacity=1,size=1).encode(x='Year:Q',y='Cement:Q')

#Field ID 8
base8 = alt.Chart(us_data).mark_circle(opacity=circle_opacity, color=colors[5], size=circle_size).encode(
    alt.X('Year:Q'),
    alt.Y('Gas Flaring:Q')
).properties(width=width, height=height)
line8 = base8.mark_line(color=colors[5], opacity=1,size=1).encode(x='Year:Q',y='Gas Flaring:Q')

#Field ID 9
base9 = alt.Chart(us_data).mark_circle(opacity=circle_opacity, color=colors[6], size=circle_size).encode(
    alt.X('Year:Q'),
    alt.Y('Per Capita:Q')
).properties(width=width, height=height)
line9 = base9.mark_line(color=colors[6], opacity=1,size=1).encode(x='Year:Q',y='Per Capita:Q')

#Field ID 10
base10 = alt.Chart(us_data).mark_circle(opacity=circle_opacity, color=colors[7], size=circle_size).encode(
    alt.X('Year:Q'),
    alt.Y('Bunker fuels (Not in Total):Q')
).properties(width=width, height=height)
line10 = base10.mark_line(color=colors[7], opacity=1,size=1).encode(x='Year:Q',y='Bunker fuels (Not in Total):Q')

alt.layer(base3,line3,base4,line4,base5,line5,base6,line6,base7,line7,base8,line8, base9,line9,base10,line10)
```

```python
#US visualizations

#scatter with trend showing correlation between liquid fuel and solid fuel
scplot = alt.Chart(us_data).mark_point().encode(
    x=alt.X('Liquid Fuel', scale=alt.Scale(domain=(0, 700000))),
    y=alt.Y('Solid Fuel',scale=alt.Scale(domain=(0, 700000)))
)
scplot 
scplot+scplot.transform_regression('Liquid Fuel', 'Solid Fuel',method="linear"
).mark_line(color="red")
```
