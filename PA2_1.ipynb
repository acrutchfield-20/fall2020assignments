```python
import numpy as np
import pandas as pd
```

```python
happy_data = pd.read_csv("input/2015.csv")
happy_data
```

```python
import altair as alt
```

```python
#binned scatter plot showing correlation between happiness
# score and family 
binplot = alt.Chart(happy_data).mark_circle().encode(
    alt.X('Family:O', bin=True),
    alt.Y('Happiness Score:O', bin=True),
    size='count()'
    
)
binplot
```

```python
#scatter plot with trendline showing correlation between happiness
# score and economy 
scat = alt.Chart(happy_data).mark_point().encode(
    x=alt.X('Economy (GDP per Capita)', scale=alt.Scale(domain=(0, 5))),
    y=alt.Y('Happiness Score',scale=alt.Scale(domain=(0, 10)))
)
scat 
scat+scat.transform_regression('Economy (GDP per Capita)', 'Happiness Score',method="linear"
).mark_line(color="red")

```

```python
#boxplot to compare more than two groups' quant range - Region and Happiness Score
alt.Chart(happy_data).mark_boxplot().encode(
    x='Region:N',
    y='Happiness Score:Q',
    color='Region:N'
).properties(width=300, height=250)
```
